```python
index = 0
def kalkulasi_z_score(row):
    if row['Jenis Kelamin'] == 'laki-laki':
        if row['Umur (bulan)'] >= 24:
            data_z_score = z_score_boy_2_5.loc[z_score_boy_2_5['Month'] == row['Umur (bulan)']]
            median = data_z_score['M'].values[0]
            
            if median > row['Tinggi Badan (cm)']:
                sd_neg_1 = data_z_score['SD1neg'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (median - sd_neg_1)
            else:
                sd_pos_1 = data_z_score['SD1'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (sd_pos_1 - median)
        else:
            data_z_score = z_score_boy_0_2.loc[z_score_boy_0_2['Month'] == row['Umur (bulan)']]
            median = data_z_score['M'].values[0]
            
            if median > row['Tinggi Badan (cm)']:
                sd_neg_1 = data_z_score['SD1neg'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (median - sd_neg_1)
            else:
                sd_pos_1 = data_z_score['SD1'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (sd_pos_1 - median)
    else:
        if row['Umur (bulan)'] >= 24:
            data_z_score = z_score_girl_2_5.loc[z_score_girl_2_5['Month'] == row['Umur (bulan)']]
            median = data_z_score['M'].values[0]
            
            if median > row['Tinggi Badan (cm)']:
                sd_neg_1 = data_z_score['SD1neg'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (median - sd_neg_1)
            else:
                sd_pos_1 = data_z_score['SD1'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (sd_pos_1 - median)
        else:
            data_z_score = z_score_girl_0_2.loc[z_score_girl_0_2['Month'] == row['Umur (bulan)']]
            median = data_z_score['M'].values[0]
            
            if median > row['Tinggi Badan (cm)']:
                sd_neg_1 = data_z_score['SD1neg'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (median - sd_neg_1)
            else:
                sd_pos_1 = data_z_score['SD1'].values[0]

                return (row['Tinggi Badan (cm)'] - median) / (sd_pos_1 - median)
```

```python
df['Z-score'] = df.apply(kalkulasi_z_score, axis=1)
```
